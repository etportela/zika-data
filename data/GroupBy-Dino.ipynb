```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('lista_dados_sage.csv',sep=';')
```

```python
import datetime
```

```python
df.head()
```

```python
df['LastDayWeek'] = pd.to_datetime((df['n_ano']).astype(str) + (df['n_semana']-1).astype(str) + "6", format="%Y%U%w")
```

```python
df['mes'] = df.LastDayWeek.dt.month
```
