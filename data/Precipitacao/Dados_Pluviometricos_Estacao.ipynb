```python
lsChuvasTot = list()

cont = 0 
for i in soupList:
    if str(type(i.body)) == "<class 'bs4.element.Tag'>":
        dados_estacao = i.body.pre.text.split('--------------------')[-1].split('\n')[1:]
        cidade = i.body.pre.text.split('--------------------')[2].split('\n')[1]
    elif str(type(i.pre)) == "<class 'bs4.element.Tag'>":
        dados_estacao = i.pre.text.split('--------------------')[-1].split('\n')[1:]
        cidade = i.pre.text.split('--------------------')[2].split('\n')[1]
        
    cabecalho = dados_estacao[0].split(';')
    #print(cont)
    #print(cabecalho[0])
    #print(cidade)
    dados = [x.split(';') for x in dados_estacao[1:]]
       
    dfChuvasTmp = pd.DataFrame(dados, columns=cabecalho)
    dfChuvasTmp['local_estacao'] = cidade
    
    print(dfChuvasTmp.head(3))
    
    lsChuvasTot.append(dfChuvasTmp)
    
    cont = cont + 1
    
dfChuvasTot = pd.concat(lsChuvasTot)
```

```python
dfChuvasTot = dfChuvasTot.dropna(subset=['Estacao'])
```

```python
dfChuvasTot #664480 
```

```python
dfChuvasTot.where(dfChuvasTot['Estacao'] == '82989').groupby(['local_estacao', 'Estacao']).Estacao.count()
```

```python
dfChuvasTot['Data_Trat'] = pd.to_datetime(dfChuvasTot['Data'], format='%d/%m/%Y')
dfChuvasTot['Ano'] = dfChuvasTot['Data_Trat'].apply(lambda x : x.year)
dfChuvasTot['Mes'] = dfChuvasTot['Data_Trat'].apply(lambda x : x.month)
```

```python

dfChuvasTot.groupby(['local_estacao']).local_estacao.count()

dfChuvasTot['local_estacao_2'] = dfChuvasTot['local_estacao'].apply(lambda x : x[20:-12])
dfChuvasTot['Municipio'] = dfChuvasTot['local_estacao_2'].apply(lambda x : x[:-5])
dfChuvasTot['UF'] = dfChuvasTot['local_estacao_2'].apply(lambda x : x[-3:])
```
