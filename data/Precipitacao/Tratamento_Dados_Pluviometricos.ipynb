### Load das bibliotecas

```python
import pandas as pd
import pickle
```

### Load do dataset

```python
# Dados pluviometricos
dados_pluviometricos = pickle.load(open("dfChuvasTot.p", "rb" ))
```

```python
dados_pluviometricos
```

### Ajustes nos dados:
* Mantém apenas os dados de 2016;
* Remove a coluna sem nome e sem dados;
* Missing Values: Substitui '' por 0;
* Reindexa o dataframe (na criação do pickle, foram acumulados os dados de todas as estações, mas o indice de cada estação foi mantido);
* Corrige os datatypes;

```python
dados_pluviometricos2016 = dados_pluviometricos[dados_pluviometricos.Ano == 2016]
dados_pluviometricos2016 = dados_pluviometricos2016.drop([''], axis=1)
dados_pluviometricos2016 = dados_pluviometricos2016.replace('', '0')
dados_pluviometricos2016 = dados_pluviometricos2016.reset_index(drop=True)
data_types = {'Evaporacao Piche': float,
              'Insolacao': float,
              'Precipitacao': float,
              'Temp Comp Media': float,
              'TempMaxima': float,
              'TempMinima': float,
              'Umidade Relativa Media': float,
              'Velocidade do Vento Media': float}
dados_pluviometricos2016 = dados_pluviometricos2016.astype(data_types)
```

```python
dados_pluviometricos2016.info()
```

```python
dados_pluviometricos2016
```

### Validação dos dados

```python
dados_pluviometricos2016.loc[(dados_pluviometricos2016['UF'].str.contains('RR'))
                              &
                             (dados_pluviometricos2016['Data_Trat'] == '2016-01-01')]
```

Cada estação coleta dados duas vezes por dia, sendo que a cada coleta, são obitidos conjuntos de variáveis diferentes.
Dessa forma, será necesário fazer um merge dos dados por estação e data de coleta

```python
aggregation_functions = {'Data': 'first',
                         'Estacao': 'first',
                         'Evaporacao Piche': 'sum',
                         'Hora': 'first',
                         'Insolacao': 'sum',
                         'Precipitacao': 'sum',
                         'Temp Comp Media': 'sum',
                         'TempMaxima': 'sum',
                         'TempMinima': 'sum',
                         'Umidade Relativa Media': 'sum',
                         'Velocidade do Vento Media': 'sum',
                         'local_estacao': 'first',
                         'Data_Trat': 'first',
                         'Ano': 'first',
                         'Mes': 'first',
                         'local_estacao_2': 'first',
                         'Municipio': 'first',
                         'UF': 'first'}
dados_pluviometricos_agg = dados_pluviometricos2016.groupby(['Data', 'Estacao']).aggregate(aggregation_functions)
dados_pluviometricos_agg = dados_pluviometricos_agg.reset_index(drop=True)
```

```python
dados_pluviometricos_agg
```

```python
dados_pluviometricos_agg.loc[(dados_pluviometricos_agg['UF'].str.contains('RR'))
                              &
                             (dados_pluviometricos_agg['Data_Trat'] == '2016-01-01')]
```

```python
dados_pluviometricos_agg.info()
```

```python
dados_pluviometricos_agg.columns
```

Descarte das colunas que não são mais necessárias

```python
dados_pluviometricos_agg2 = dados_pluviometricos_agg.drop(['Data', 'Estacao', 'Hora', 'local_estacao', 'local_estacao_2', 'Municipio'], axis=1)
```

```python
dados_pluviometricos_agg2.info()
```

```python
dados_pluviometricos_agg2
```

Faz um novo merge, dessa vez, para gerar dados mensais por Ano, Mes e UF

```python
aggregation_functions = {'Evaporacao Piche': 'mean',
                         'Insolacao': 'mean',
                         'Precipitacao': 'mean',
                         'Temp Comp Media': 'mean',
                         'TempMaxima': 'mean',
                         'TempMinima': 'mean',
                         'Umidade Relativa Media': 'mean',
                         'Velocidade do Vento Media': 'mean',
                         'Data_Trat': 'first',
                         'Ano': 'first',
                         'Mes': 'first',
                         'UF': 'first'}
dados_pluviometricos_agg3 = dados_pluviometricos_agg2.groupby(['Ano', 'Mes', 'UF']).aggregate(aggregation_functions)
dados_pluviometricos_agg3 = dados_pluviometricos_agg3.drop(['Data_Trat'], axis=1)
dados_pluviometricos_agg3 = dados_pluviometricos_agg3.reset_index(drop=True)
```

```python
dados_pluviometricos_agg3.info()
```

```python
dados_pluviometricos_agg3
```

### Salvando os dados

```python
# csv
dados_pluviometricos_agg3.to_csv('dfChuvasFinal.csv', index=False, sep=';')

# pickle
pickle.dump(dados_pluviometricos_agg3,open('dfChuvasFinal.p', "wb"))
```
