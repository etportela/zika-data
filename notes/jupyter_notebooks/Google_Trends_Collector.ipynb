## Busca de interesse por região e período

```python
pytrends = TrendReq(hl='en-US', tz=360)

kw_list = ["Zika Virus"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='BR', gprop='')
```

```python
# Interest by Region
df = pytrends.interest_by_region(resolution='DMA', inc_low_vol=True, inc_geo_code=True)
df
```
