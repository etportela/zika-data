# HISTOGRAMA

O histograma é utilizado para resumir a distribuição de um conjunto de dados.

```python
import pandas  as pd
import numpy   as np
import seaborn as sns
import matplotlib.pyplot as plt

#dados x
x1=np.array([dataset.n_semana])     
x2=np.array([dataset.n_ano])                     
x3=np.array([dataset.total_acumulado]).reshape(-1, 1)            
x4=np.array([dataset.total_investigacao]).reshape(-1, 1)         
x5=np.array([dataset.total_confirmado]).reshape(-1, 1)           
x6=np.array([dataset.total_descartado]).reshape(-1, 1) 

fig,ax = plt.subplots()
ax.hist([x1,x2,x3,x4,x5,x6])
ax.set(title='DISTRIBUIÇÂO NORMAL')

plt.xlabel("ZIKA VIRUS")
plt.ylabel("FREQUENCIA")

#legendas
legend = ['n_semana', 'n_ano','total_acumulado','total_investigacao',
          'total_confirmado','total_descartado']
plt.legend(legend)
```

# REGRESSÃO

### Modelo de Regressão KNN e Linear
#### KNN (K-Nearest Neighbors)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
```

```python
zv = np.array([dataset.n_semana]).reshape(-1, 1)
semanas = np.array([dataset.total_confirmado]).reshape(-1, 1)

zv_teste = np.array([10,30,50]).reshape(-1,1)
semanas_teste = np.array([50,150,175]).reshape(-1,1)

```

```python
plt.scatter(zv, semanas)
plt.grid(True)
plt.title("SEMANAS x ZIKA VIRUS")
plt.ylabel("ZIKA VIRUS")
plt.xlabel("SEMANAS")
plt.show()
```

```python
#Importação do modelo KNN e Linear
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
```

```python
# Criando os modelos
knn_1 = KNeighborsRegressor(n_neighbors = 1).fit(zv, semanas)
knn_3 = KNeighborsRegressor(n_neighbors = 3).fit(zv, semanas)
knn_5 = KNeighborsRegressor(n_neighbors = 5).fit(zv, semanas)
linear = LinearRegression().fit(zv, semanas)
```

```python
# Exibindo os gráficos
# Exibindo as linhas de predição
#legendas = ["Linha de prediçao", "Treino", "Teste"]
figura, eixos = plt.subplots(2, 2, figsize=(10,10))


# Código para criar a linha de predição dos dados de treino
# onde é criado 500 valores igualmente espaçados entre 1.45 e 1.95
linha = np.linspace(1.45, 1.95, 1000).reshape(-1, 1)

#Knn 1
eixo = eixos[0,0]
eixo.grid(True)
eixo.plot(linha, knn_1.predict(linha))
eixo.plot(zv, semanas, "^", markersize=5)
eixo.plot(zv_teste, semanas_teste, "v", markersize=8)
eixo.set_title("KNN {}\n Treino: {:.2f} - Teste: {:.2f}".format(1, knn_1.score(zv, semanas), knn_1.score(zv_teste, semanas_teste)))

#Knn 3
eixo = eixos[0,1]
eixo.grid(True)
eixo.plot(linha, knn_3.predict(linha))
eixo.plot(zv, semanas, "^", markersize=5)
eixo.plot(zv_teste, semanas_teste, "v", markersize=8)
eixo.set_title("KNN {}\n Treino: {:.2f} - Teste: {:.2f}".format(3, knn_3.score(zv, semanas), knn_3.score(zv_teste, semanas_teste)))

#Knn 5
eixo = eixos[1,0]
eixo.grid(True)
eixo.plot(linha, knn_5.predict(linha))
eixo.plot(zv, semanas, "^", markersize=5)
eixo.plot(zv_teste, semanas_teste, "v", markersize=8)
eixo.set_title("KNN {}\n Treino: {:.2f} - Teste: {:.2f}".format(5, knn_5.score(zv, semanas), knn_5.score(zv_teste, semanas_teste)))

#Linear
eixo = eixos[1,1]
eixo.grid(True)
eixo.plot(linha, linear.predict(linha))
eixo.plot(zv, semanas, "^", markersize=5)
eixo.plot(zv_teste, semanas_teste, "v", markersize=8)
eixo.set_title("Modelo Linear\n Treino: {:.2f} - Teste: {:.2f}".format(linear.score(zv, semanas), linear.score(zv_teste, semanas_teste)))

plt.show()
```
